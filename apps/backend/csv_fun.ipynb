```python
import pandas as pd

# Load the CSV file
df = pd.read_csv('anonymized_resumes.csv')

# Display basic information about the dataframe
print("DataFrame Info:")
print(f"Shape: {df.shape}")
print("\nColumns:")
print(df.columns.tolist())

# Display the first few rows
print("\nFirst 5 rows:")
display(df.head())

# Check for missing values
print("\nMissing values:")
print(df.isnull().sum())

# Basic statistics for numeric columns
print("\nBasic statistics:")
display(df.describe(include='all'))

# Get the number of rows in the dataframe
num_rows = len(df)
print(f"\nNumber of rows in the dataframe: {num_rows}")

```

```python
# Get the first row in the dataframe
first_row = df.iloc[0]
print("\nFirst row in the dataframe:")
print(first_row)

# Print the education column of the first row
print("\nEducation column of the first row:")
print(first_row['education'])

```
